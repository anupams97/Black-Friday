# Black Friday 

Importing required libraries and data preprocessing.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Reading the testing and training data set.

```python
df1 = pd.read_csv('train.csv') 
df2 = pd.read_csv('test.csv')
```

```python
df1.head() #Training dataset

```

```python
df2.head() #test dataset
```

Concatinating the training and test dataset.


```python
frames= [df1,df2]
df = pd.concat(frames , ignore_index= 'True')
df = df.reindex_axis(df1.columns, axis=1)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

replacing the missing values with 999.

```python
df = df.fillna(999)
```

```python
df.describe(include='all')
```
